```c++
#include <iostream>
#include <stdlib.h>
#include <string>
#include <time.h>
#include <unordered_map>

using std::cout;
```

```c++
srand(time(NULL));
```

```c++
int randint(int lowerbound, int upperbound) {
    srand(time(NULL));
    return rand() % (upperbound - lowerbound + 1) + lowerbound;
}
```

```c++
const int SAMPLE_SIZE = 10000;
std::unordered_map<std::string, int> histogram;
int lb = 1, ub = 4;

for (int i = 0; i < SAMPLE_SIZE; ++i) {
    int rint = randint(lb, ub);
    if (histogram.find(rint) == histogram.end()) {
        histogram.insert(rint, 1);
    } else {
        histogram[rint]++;
    }
}
```
