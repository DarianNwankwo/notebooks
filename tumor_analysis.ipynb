```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

tumor_img = Image.open("./images/tumor_fluor.jpeg")
```

```python
tumor_img
```

```python
def isBlackish(pixel, threshold=0):
    red, green, blue = pixel
    return red < threshold and green < threshold and blue < threshold
```

```python
thresholds = range(0, 255, 3)
black_counts = []
total_pixels = tumor_img.width * tumor_img.height

for T in thresholds:
    cur_count = 0
    for x in range(tumor_img.width):
        for y in range(tumor_img.height):
            cur_pixel = tumor_img.getpixel((x, y))
            if isBlackish(cur_pixel, threshold=T):
                cur_count += 1
    black_counts.append(cur_count)
# blackness_ratio = black_count / total_pixels
```

```python
plt.axhline(y=total_pixels, color='r', linestyle='-')
plt.axhline(y=total_pixels/2, color='r', linestyle='-')
plt.plot(black_counts)
```

```python
[(b/total_pixels, ndx) for ndx, b in enumerate(black_counts)]
```
